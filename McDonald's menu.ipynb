```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("\n=== DESCRIPTIVE STATISTICS ===")
desc_stats = df[['Calories', 'Total Fat', 'Sodium', 'Carbohydrates', 'Protein']].describe()
print(desc_stats)
```

```python
print("\nHighest Calorie Items:")
print(df.nlargest(5, 'Calories')[['Category', 'Item', 'Calories']])
```

```python
print("\nLowest Calorie Items (excluding zero-calorie beverages):")
print(df[df['Calories'] > 0].nsmallest(5, 'Calories')[['Category', 'Item', 'Calories']])
```

```python
print("\n=== CATEGORY ANALYSIS ===")
category_stats = df.groupby('Category').agg({
    'Calories': ['mean', 'median', 'min', 'max'],
    'Total Fat': 'mean',
    'Protein': 'mean'
}).sort_values(('Calories', 'mean'), ascending=False)
print(category_stats)
```

```python
print("\n=== HEALTHIEST OPTIONS ===")
healthy_criteria = (df['Calories'] < 400) & (df['Protein'] > 15) & (df['Total Fat'] < 20)
print("High-protein, moderate-calorie options:")
print(df[healthy_criteria][['Category', 'Item', 'Calories', 'Protein', 'Total Fat']].sort_values('Protein', ascending=False))
```

```python
plt.figure(figsize=(14, 8))

category_order = df.groupby('Category')['Calories'].median().sort_values(ascending=False).index

sns.boxplot(
    data=df,
    x='Category',
    y='Calories',
    hue='Category',               # Explicitly assign hue
    order=category_order,
    palette='viridis',
    width=0.6,
    linewidth=1.5,
    showfliers=False,
    legend=False                  # Turn off legend (optional here)
)

sns.stripplot(
    data=df,
    x='Category',
    y='Calories',
    order=category_order,
    color='black',
    size=4,
    alpha=0.3,
    jitter=True
)

plt.title('Calorie Distribution by Menu Category', fontsize=16, fontweight='bold')
plt.xlabel('Menu Category', fontsize=12)
plt.ylabel('Calories', fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Ensure the layout works with other subplots
plt.subplot(2, 2, 2)

# Step 1: Compute mean of macronutrients per category
macros = df.groupby('Category')[['Total Fat', 'Carbohydrates', 'Protein']].mean()

# Step 2: Convert to percentage composition
macros_pct = macros.div(macros.sum(axis=1), axis=0) * 100

# Step 3: Sort categories according to median calorie order (from earlier boxplot)
macros_pct = macros_pct.loc[category_order]

# Step 4: Plot the stacked bar chart
macros_pct.plot(
    kind='bar',
    stacked=True,
    color=['#FF9F1C', '#2EC4B6', '#E71D36'],  # Fat, Carbs, Protein
    width=0.7,
    edgecolor='black',
    linewidth=0.6,
    ax=plt.gca()  # Use current subplot
)

# Step 5: Title and labels
plt.title(' Macronutrient Composition by Category', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Macronutrient % Composition', fontsize=12)
plt.xlabel('Menu Category', fontsize=12)

# Improve ticks and layout
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 100)
plt.legend(
    title='Macronutrient',
    title_fontsize=10,
    fontsize=9,
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)
plt.grid(axis='y', linestyle='--', alpha=0.4)
sns.despine()
```

```python
# Top 10 High Sugar Items
plt.subplot(2, 2, 4)
top_sugar = df.nlargest(10, 'Sugars')[['Item', 'Sugars']]
sns.barplot(data=top_sugar, y='Item', x='Sugars')
plt.title('Top 10 Highest Sugar Items')

plt.tight_layout()
plt.show()
```

```python
# Heatmap of Nutritional Values
heatmap_data = df.pivot_table(index='Category', 
                             values=['Calories', 'Total Fat', 'Sodium', 'Protein'], 
                             aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlOrRd")
plt.title('Average Nutritional Values by Category')
plt.show()
```
